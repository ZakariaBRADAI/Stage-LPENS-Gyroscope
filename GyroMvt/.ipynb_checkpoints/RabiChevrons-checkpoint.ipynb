# II- Chevrons de Rabi "quantiques" : 

```python
def Rabi_Ideal(t_burst, delta, Omega_R):
    '''Chevrons de Rabi id√©aux.'''
    tab_Rabi_th = np.zeros((len(delta), len(t_burst)))
    for i in range(len(delta)):
        for j in range(len(t_burst)):
            pulsation = np.sqrt(delta[i]**2 + (Omega_R**2)) #/ 2
            num = (Omega_R**2) * np.sin( pulsation * t_burst[j] )**2
            den = (Omega_R**2) + delta[i]**2
            tab_Rabi_th[i, j] = num / den
    return tab_Rabi_th
```

```python
Rabi_spin = Rabi_Ideal(t_burst, delta, omega_R_th)

plt.figure(figsize=(4, 3), dpi = 300)
plt.imshow(Rabi_spin, origin = "lower", extent=[0, 20, 0.75, 1.35], aspect="auto")
plt.xlabel("Dur√©e de l'excitation [s]")
plt.ylabel("Fr√©quence d'excitation [Hz]")
plt.savefig('Figures/Rabi/Rabi_Ideal_Th.png')
plt.colorbar()
plt.show()
```

# III- Chevrons de Rabi avec amplitude modul√©e :

```python
def Rabi_Freq_Modified(delta, Omega_R, Omega_L):
    return Omega_R * (1 + delta / Omega_L)**2


def Rabi_Assym(t_burst, delta, Omega_R, Omega_L):
    '''Chevrons de Rabi assym√©triques.'''
    tab_Rabi_th = np.zeros((len(delta), len(t_burst)))
    for i in range(len(delta)):
        for j in range(len(t_burst)):
            Omega_R_mod = Rabi_Freq_Modified(delta[i], Omega_R, Omega_L)
            pulsation =  np.sqrt(delta[i]**2 + Omega_R_mod**2) # / 2
            num = (Omega_R_mod**2) * np.sin( pulsation * t_burst[j] )**2
            den = Omega_R_mod**2 + delta[i]**2
            tab_Rabi_th[i, j] = num / den
    return tab_Rabi_th
```

```python
Rabi_assym = Rabi_Assym(t_burst, delta, omega_R_th, omega_L_th)


plt.figure(figsize=(4, 3), dpi = 300)
plt.imshow(Rabi_assym, origin = "lower", extent=[0, 20, 0.75, 1.35], aspect="auto")
plt.xlabel("Dur√©e de l'excitation [s]")
plt.ylabel("Fr√©quence d'excitation [Hz]")
plt.colorbar()
plt.savefig('Figures/Rabi/Rabi_Assym_Th.png')
plt.show()
```
